# 10. **BONUS QUESTION: Out of the 5 sections, 2 sections have a high correlation. Find those sections and prove by a valid plot/graph/chart/table with a proper reasoing statement. 

```python
df9=df2
import math as m
import matplotlib.pyplot as plt
ls=['Section1','Section2','Section3','Section4','Section5']
for i in range(len(ls)):
    for j in range(i,len(ls)):
        if i!=j:
            df9['Xterms']=(df9[ls[i]]-df9[ls[i]].mean())**2
            df9['Yterms']=(df9[ls[j]]-df9[ls[j]].mean())**2
            df9['product of diff']=(df9[ls[i]]-df9[ls[i]].mean())*(df9[ls[j]]-df9[ls[j]].mean())


            print('The correlation of {},{} is{}'.format(ls[i],ls[j],df9['product of diff'].sum()/(m.sqrt(df9['Xterms'].sum())*m.sqrt(df9['Yterms'].sum()))))
            

plt.figure(figsize=[25,25],facecolor='yellow')
plt.style.use('fivethirtyeight')
plt.subplot(4,4,1)
plt.xlabel('section1')
plt.ylabel('section2')
plt.title('section2 vs section1')
plt.scatter(df9['Section1'],df9['Section2'])
plt.subplot(4,4,2)
plt.xlabel('section1')
plt.ylabel('section3')
plt.title('section3 vs section1')
plt.scatter(df9['Section1'],df9['Section3'])

plt.subplot(4,4,3)
plt.xlabel('section1')
plt.ylabel('section4')
plt.title('section4 vs section1')
plt.scatter(df9['Section1'],df9['Section4'])

plt.subplot(4,4,4)
plt.xlabel('section1')
plt.ylabel('section5')
plt.title('section5 vs section1')
plt.scatter(df9['Section1'],df9['Section5'])

plt.subplot(4,4,5)
plt.xlabel('section2')
plt.ylabel('section3')
plt.title('section3 vs section2')
plt.scatter(df9['Section2'],df9['Section3'])

plt.subplot(4,4,6)
plt.xlabel('section2')
plt.ylabel('section4')
plt.title('section4 vs section2')
plt.scatter(df9['Section2'],df9['Section4'])

plt.subplot(4,4,7)
plt.xlabel('section2')
plt.ylabel('section5')
plt.title('section5 vs section2')
plt.scatter(df9['Section2'],df9['Section5'])

plt.subplot(4,4,8)
plt.xlabel('section3')
plt.ylabel('section4')
plt.title('section4 vs section3')
plt.scatter(df9['Section3'],df9['Section4'])

plt.subplot(4,4,9)
plt.xlabel('section3')
plt.ylabel('section5')
plt.title('section5 vs section3')
plt.scatter(df9['Section3'],df9['Section5'])

plt.subplot(4,4,10)
plt.xlabel('section4')
plt.ylabel('section5')
plt.title('section5 vs section4')
plt.scatter(df9['Section4'],df9['Section5'])
plt.show()


print('From graph and formula it is clear that "section1 and section3" has the highest correlation because the graph is comparatively having more postive slope than the other graphs')

    
```

# 14. Print the names of students in every section who have secured the maximum Correct/Total ratio. If there are multiple topper students in any section, print the count and names of all of them.

```python
marks={'ratio1':'section1','ratio2':'section 2','ratio3':'section 3','ratio4':'section 4','ratio5':'section 5'}
df7['ratio1']=df7['Marks 1']/df7['Total Score']
df7['ratio2']=df7['Marks 2']/df7['Total Score']
df7['ratio3']=df7['Marks 3']/df7['Total Score']
df7['ratio4']=df7['Marks 4']/df7['Total Score']
df7['ratio5']=df7['Marks 5']/df7['Total Score']
for i in marks.keys():
    y=df7[i].max()
    print('The Topper of {} (with maximum Correct/Total ratio) is {}'.format(marks[i],df7[df7[i]==y].Name))
    print('There are total {} such toppers in {}'.format(len(df7[df7[i]==y]),marks[i]))
    print('*******************')
```

# 15. Find out if there is any student who is topper in both above questions (section wise). Such students will be called scholars. Prints the names of all scholars

```python
marks1={'Marks 1':'section 1','Marks 2':'section 2','Marks 3':'section 3','Marks 4':'section 4','Marks 5':'section 5'}
marks2={'ratio1':'section1','ratio2':'section 2','ratio3':'section 3','ratio4':'section 4','ratio5':'section 5'}
for i,j in zip(marks1.keys(),marks2.keys()):
    y=df7[i].max() 
    z= df7[j].max()
    s1=set(df7[df7[i]==y].Name)
    s2=set(df7[df7[j]==z].Name)
    print('Scholar of {} is {}'.format(marks2[j],s1.intersection(s2)))
    print('#################')
```
